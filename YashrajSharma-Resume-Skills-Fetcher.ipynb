## **1) Libraries**

```python
import pandas as pd
import json
```

## **2) Initialization**

```python
BASEDIR = "/content/drive/My Drive/Python Live/Resume-Fetcher/"
```

```python
# Expecting skill sets from resume
skill_sets = ['management', 'problem solving', 'sql']

# Initializing default dataframe to save the results
df = pd.DataFrame(columns=['Skills','Yes or No'])
df['Skills']=['Management','problem solving','SQL']
df['Yes or No'].fillna('No',inplace=True)
```

## **3) Parsing JSON**

```python
# Opening JSON file
f = open(BASEDIR+'sample.json',)
data = json.load(f) 
```

## **4) Functions to fetch skills & saving the resul in dataframe**

```python
# Function to update Flag in DataFrame
def set_df_flag(type, isAvailable):
  df.loc[df[df['Skills'].str.lower() == type].index, 'Yes or No'] = isAvailable

# Function to fetch the desired skills from resume
def skillsFetcher(data_):
    #TEST data
    #data.update({"Resume":{"StructuredResume":{"Competency":[{"skillAliasArray":["management"]}]}}})

    for competency in data_['Resume']['StructuredResume']['Competency']:
        if 'skillAliasArray' in competency:
            for skillsArray in competency['skillAliasArray']:
                if(skillsArray.lower() in skill_sets):
                  set_df_flag(skillsArray.lower(),'Yes')
        else:
            if(competency['skillName'].lower() in skill_sets):
                set_df_flag(competency['skillName'].lower(),'Yes')
```

```python
skillsFetcher(data)
```
